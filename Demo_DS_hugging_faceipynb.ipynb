```python
from huggingface_hub import hf_hub_download, list_models

# Download a file from the Hugging Face Hub
# This example downloads the config.json file from the "bert-base-uncased" model
try:
    file_path = hf_hub_download(repo_id="bert-base-uncased", filename="config.json")
    print(f"Downloaded file to: {file_path}")
except Exception as e:
    print(f"Error downloading file: {e}")


# List models on the Hugging Face Hub
# This example lists the first 10 models
try:
    models = list_models(limit=10)
    print("\nFirst 10 models on the Hub:")
    for model in models:
        print(model.modelId)
except Exception as e:
    print(f"Error listing models: {e}")
```
