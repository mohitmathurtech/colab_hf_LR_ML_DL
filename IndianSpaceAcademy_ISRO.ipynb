```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load GeoJSON
gdf = gpd.read_file("/content/drive/MyDrive/GEE_P1_WaterNDWI/water_polygons.geojson")

# Show first rows
print(gdf.head())

# Ensure theres an ID column (if not, create one)
if "id" not in gdf.columns:
    gdf["id"] = range(1, len(gdf) + 1)

# If area not present, calculate in sq. km
if "area_sq_km" not in gdf.columns:
    gdf["area_sq_km"] = gdf.geometry.area / 1e6

# Plot water bodies
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color="skyblue", edgecolor="black")



# Improve map appearance
ax.set_title("Water Bodies Detected (NDWI)", fontsize=14, weight="bold")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
plt.axis("equal")

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cx # Library for adding basemaps

# Load the GeoJSON file
gdf = gpd.read_file("/content/drive/MyDrive/GEE_P1_WaterNDWI/water_polygons.geojson")

# Reproject to a Web Mercator CRS (required for contextily)
gdf = gdf.to_crs(epsg=3857)

# Create the plot
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the water bodies
gdf.plot(ax=ax, color="blue", edgecolor="black", label="Water Bodies")

# Add a basemap from a provider like OpenStreetMap
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik)

# Add a legend to explain the colors
ax.legend(loc='upper right', title="Legend")

# Improve map appearance
ax.set_title("Water Bodies Detected (NDWI) with Basemap", fontsize=16, weight="bold")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# Remove x and y axis ticks for a cleaner map
ax.set_xticks([])
ax.set_yticks([])

plt.show()
```
